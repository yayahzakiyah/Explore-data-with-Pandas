**3. The name of the director and total gross of the films that have been directed.**

To more easily retrieve the data director, it is necessary to separate the Stars column.

```python
#Split the STARS column into Director and Start columns separated by |
movies[['Director', 'Stars']] = movies['STARS'].str.rsplit('|', expand=True)
```

```python
movies
```

In the dataset above, there are still names of stars in the Director column. so it needs to be moved.

Create a function that combines the Director and Stars columns into a new stars column but with the condition that the Stars column has no value or None.

```python
def stars(row):
    director = row['Director']
    Stars = row['Stars']

    if pd.isnull(Stars):
        return director
    else:
        return Stars

movies['stars'] = movies.apply(stars, axis=1)
```

```python
movies
```

Then create a new function where the Director and stars columns are combined but for the same row value replace None into new column director

```python
def director(row):
    Director = row['Director']
    Stars = row['stars']

    if Director == Stars:
        return None
    else:
        return Director

movies['director'] = movies.apply(director, axis=1)
```

```python
movies
```

```python
#Delete Directur and Stars columns
movies = movies.drop(['Director', 'Stars'], axis = 1)
```

```python
#Delete Director and Directors words in director column
movies['director'] = movies['director'].replace({'Director:':''}, regex=True)
movies['director'] = movies['director'].replace({'Directors:':''}, regex=True)
```

```python
movies.head(50)
```

```python
movies.info()
```

For sum gross, gros have to Dtype float not object

```python
movies['Gross'] = movies['Gross'].str.replace(r'\D', '')
```

```python
movies[movies['Gross'].notnull()]
```

```python
#Convert to numeric
movies['Gross'] = pd.to_numeric(movies['Gross'])
movies.info()
```

```python
#Director column is grouped then add Gross
name_director = movies.groupby(['director'])['Gross'].agg('sum').reset_index(name='Amount')
name_director.sort_values('Amount', ascending=False).head(30)
```
